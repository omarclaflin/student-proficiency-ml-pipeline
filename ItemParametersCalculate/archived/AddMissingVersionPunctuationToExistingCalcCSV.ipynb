```python
def add_decimal(number):
    number_str = str(number)
    return float(number_str[0] + '.' + number_str[1:])

# Apply the function to the 'version' column
df['version'] = df['version'].apply(add_decimal)
df
```

```python
df.to_csv('ItemParameterEstimates.csv',index=False)
```
