```python
import pandas as pd

def clean_csv(input_file, output_file, columns_to_keep):
    """
    Read a CSV file, keep values only in specified columns and save to a new CSV file.
    All other columns will have their values deleted but column names preserved.
    
    Args:
        input_file (str): Path to the input CSV file
        output_file (str): Path to the output CSV file
        columns_to_keep (list): List of column names whose values should be preserved
    """
    try:
        # Read the CSV file
        df = pd.read_csv(input_file)
        
        # Print original data info
        print(f"Original CSV has {len(df)} rows and {len(df.columns)} columns")
        print(f"Columns in the original file: {list(df.columns)}")
        
        # Create a copy of the dataframe with all values set to empty strings
        df_cleaned = df.copy()
        
        # For each column not in columns_to_keep, set all values to empty string
        for column in df.columns:
            if column not in columns_to_keep:
                df_cleaned[column] = ""
        
        # Save the cleaned dataframe to a new CSV file
        df_cleaned.to_csv(output_file, index=False)
        
        # Print info about the cleaned data
        print(f"Cleaned CSV saved to {output_file}")
        print(f"Values preserved only in columns: {columns_to_keep}")
        
    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    # Input and output file paths
    input_file = "Math_ItemParameters_v5_empty.csv"
    output_file = "Math_ItemParameters_v5_fullyempty.csv"
    
    # List of columns whose values should be preserved
    columns_to_keep = ["question_id", "skill_id"]
    
    # Process the file
    clean_csv(input_file, output_file, columns_to_keep)
```
