```python
import subprocess

def list_docker_images():
    try:
        result = subprocess.run(["docker", "images"], capture_output=True, text=True)
        if result.returncode == 0:
            return result.stdout
        else:
            return f"Error: {result.stderr}"
    except FileNotFoundError:
        return "Docker CLI not found. Make sure Docker is installed."

print('LOCAL Docker images: ')
print(list_docker_images())

import docker

def get_image_id(image_name):
    try:
        client = docker.from_env()
        images = client.images.list(filters={"reference": image_name})
        if images:
            return images[0].id
        else:
            return f"No image found with the name '{image_name}'."
    except docker.errors.DockerException as e:
        return f"Error: {e}"

image_name = current_model
local_image_id = get_image_id(image_name)
print(f"Image ID for '{image_name}': {local_image_id}")
```

```python
import subprocess

def authenticate_ecr(region, aws_account_id):
    cmd = f"aws ecr get-login-password --region {region} | docker login --username AWS --password-stdin {aws_account_id}.dkr.ecr.{region}.amazonaws.com"
    subprocess.run(cmd, shell=True, check=True)

# Example usage:
region = "us-east-1"
aws_account_id = "[REDACTED]"
authenticate_ecr(region, aws_account_id)
```

```python
#build and push to repo

import subprocess

def tag_and_push_image(repository_uri, local_image_id, tag='latest'):
    image_uri = f"{repository_uri}:{tag}"
    subprocess.run(['docker', 'tag', local_image_id, image_uri])
    subprocess.run(['docker', 'push', image_uri])

# Replace with your ECR repository URI and local image ID
ecr_repository_uri = '[REDACTED].dkr.ecr.us-east-1.amazonaws.com/[REDACTED]-ml'
tag_and_push_image(ecr_repository_uri, local_image_id)
```
