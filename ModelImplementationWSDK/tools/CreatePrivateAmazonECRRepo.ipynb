```python
import boto3

def create_ecr_repository(repo_name, region_name):
    try:
        ecr_client = boto3.client('ecr', region_name=region_name)  # Replace 'your-region' with your desired AWS region
        response = ecr_client.create_repository(repositoryName=repo_name)
        print(f"Repository '{repo_name}' created successfully!")
        print(f"Repository URI: {response['repository']['repositoryUri']}")
    except Exception as e:
        print(f"Error creating repository: {e}")

if __name__ == "__main__":
    repository_name = "[REDACTED]-ml"  # Replace with your desired repository name
    region_name='us-east-1'
    create_ecr_repository(repository_name, region_name)
```
