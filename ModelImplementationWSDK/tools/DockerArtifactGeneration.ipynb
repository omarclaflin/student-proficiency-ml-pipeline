```python
#build docker

def generate_dockerfile(current_model: str, 
                        base_image: str = "jupyter/scipy-notebook") -> str:
    """
    Generates a Dockerfile for deploying a machine learning model.

    Args:
        model_filename (str): The filename of the trained model (e.g., "final_model.joblib").
        app_filename (str): The filename of your Python script that uses the model (e.g., "my_app.py").
        base_image (str, optional): Base Docker image to use. Defaults to "python:3.8-slim-buster".

    Returns:
        str: The content of the Dockerfile.
    """
    dockerfile = f"""\
# Use an official Python runtime as the base image
FROM {base_image}

# Set the working directory inside the container
WORKDIR /usr/src/app/run

# Copy the model file into the container
COPY {current_model}/model_inference_API.py run/inference.py
COPY {current_model}/dockerfile .
COPY {current_model}/requirements.txt .

# Install any additional dependencies (if needed)
RUN pip install -r requirements.txt

# Specify the command to run when the container starts
CMD ["python", "-u", "inference.py"]
"""

    return dockerfile

# Example usage:
current_model = 'model05292024'
dockerfile_content = generate_dockerfile(current_model)

# Print the generated Dockerfile content
print(dockerfile_content)

```

```python
# Define the requirements.txt
# my_list = ['boto3==1.34.32',
#            'snowflake-connector-python==3.7.1',
#            'joblib==1.3.2',
#            'flask==3.0.2', 
#            'waitress==3.0.0']

my_list = ['flask','waitress','joblib', 'sagemaker']

# Create a string with each element on a separate line
requirements_content = "\n".join(my_list)

# Print the generated requirements content
print(requirements_content)
```

```python
#check if folder exists
import os
def create_folder_if_not_exists(folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Folder '{folder_path}' created.")
    else:
        print(f"Folder '{folder_path}' already exists.")

create_folder_if_not_exists(current_model)


# write out dockerfile
with open(current_model+'/'+"dockerfile", "w") as text_file:
    text_file.write(dockerfile_content)

print("String written to "+current_model+'/'+"dockerfile")

# write out requiements
with open(current_model+'/'+"requirements.txt", "w") as text_file:
    text_file.write(requirements_content)

print("String written to "+current_model+'/'+"requirements.txt")

```
