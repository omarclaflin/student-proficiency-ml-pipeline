```python
import xgboost as xgb

# Load the model
model = xgb.Booster(model_file='models/xgb_model.json')

# Print feature importances
print('\n'.join([f"{feature}: {importance:.4f}" for feature, importance in sorted(zip(model.feature_names, model.get_score(importance_type='gain').values()), key=lambda x: x[1], reverse=True)]))
```

```python
# Print feature importances
for feature, importance in sorted(zip(model.feature_names, model.get_score(importance_type='gain').values()), key=lambda x: x[1], reverse=True):
    if 'OCCURREDAT_DIFF' in feature:
        print(f"{feature}: {importance:.4f}")
```
